```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# Load embedding model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Load Hugging Face LLM for generation (DistilGPT2 model)
generator = pipeline(
    "text-generation",
    model="distilgpt2",
    device_map="auto"
)
```

```python
documents = [
    "LangChain is a framework that helps developers build applications using language models.",
    "FAISS is a vector search library developed by Facebook AI Research.",
    "Mistral 7B is an open-source language model designed for high-quality inference.",
    "Hugging Face Transformers provide thousands of pre-trained models for NLP."
]
```

```python
doc_embeddings = embedder.encode(documents).astype('float32')

# Build index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
def retrieve_context(query, k=2):
    query_embedding = embedder.encode([query]).astype('float32')
    distances, indices = index.search(query_embedding, k)
    return [documents[i] for i in indices[0]]

def generate_response(query):
    context = "\n".join(retrieve_context(query))
    prompt = f"""### Instruction:
Use the following context to answer the question.

Context:
{context}

Question:
{query}

### Response:"""

    output = generator(prompt, max_new_tokens=150, do_sample=True, temperature=0.7)
    return output[0]['generated_text'].split("### Response:")[-1].strip()
```

```python
query = "What is FAISS used for?"
print("🧠 Query:", query)
print("🤖 Answer:", generate_response(query))
```
